```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 10               # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
# 복합그래프 - 여러개 그래프
# 4개 그래프를 혼합해서 사용
fig, axs = plt.subplots(2,2,figsize=(15,13))

# 메인타이틀 생성
# axs[0,0], [0,1], [1,0], [1,1]
fig.suptitle('4개 복합 그래프')

# 4개 그래프 - 1번째 
axs[0,0].bar(df['이름'], df['국어'], label='국어')
axs[0,0].set_title('1. 국어 점수 그래프')
# x축, y축 이름 설정
axs[0,0].set(xlabel='이름', ylabel='점수')
# 배경색
axs[0,0].set_facecolor('lightyellow')
axs[0,0].legend(loc='upper center')
for i in range(len(df['국어'])):
  axs[0,0].text(df['이름'][i], df['국어'][i]+2,df['국어'][i], ha='center')

# 2번째 선그래프
axs[0,1].set_title('2. 영어, 수학 그래프')
axs[0,1].set(xlabel='이름', ylabel='점수')
axs[0,1].plot(df['이름'], df['영어'], label='영어')
axs[0,1].plot(df['이름'], df['수학'], label='수학')
axs[0,1].legend()
for i in range(len(df['영어'])):
  axs[0,1].text( df['이름'][i],df['영어'][i],df['영어'][i], ha='center')
  axs[0,1].text( df['이름'][i],df['수학'][i],df['수학'][i], ha='center')



## 3번째 : 가로바 그래프
axs[1,0].set_title('3. 키 그래프')
axs[1,0].set_xlim(150,210)
axs[1,0].barh(df['이름'], df['키'])
for i in range(len(df['키'])):
  axs[1,0].text( df['키'][i]+3,df['이름'][i],df['키'][i], ha='center')


# 4번째 : 선그래프
axs[1,1].set_title('4. 사회 그래프')
axs[1,1].plot(df['이름'],df['사회'], color='g', alpha=0.5)

for i in range(len(df['사회'])):
  axs[1,1].text(df['이름'][i], df['사회'][i]+3,df['사회'][i], ha='center')

plt.show()
```
