```python
# values = [30,25,20,13,10,2]    # 100 퍼센트
values = [1,2,2,1,1,1]    # 100 퍼센트
explode = [0.3,0.1,0,0,0,0]
labels = ['python','java','javascript','oracle','html', 'etc']
colors = ['y','g','peachpuff','bisque','plum','m']

plt.title('프로그램 사용률')

# pie 함수(데이터값, labels = 글자출력)
# explode = 원그래프 분리 / counterclock=False : 원 방향을 시계방향으로 / startangle= 시작점 지정
# 데이터값 출력 : autopct='%.0f%%'
plt.pie(values, labels=labels, explode=explode, autopct='%.0f%%', counterclock=False, startangle=90, colors=colors)
# plt.legend(loc='upper right')
plt.legend(loc=(1.2,0.3), title='언어별 선호도')
plt.show()
```
