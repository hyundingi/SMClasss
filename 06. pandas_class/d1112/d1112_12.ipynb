```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

* drink.csv 불러오기
* total_servings 컬럼을 모든 servings 합계로 추가
* 나라별 막대그래프 생성 country

```python
df = pd.read_csv('drinks.csv', sep=',')
df['total_servings'] = df['beer_servings']+df['spirit_servings']+df['wine_servings']

index = np.arange(df.shape[0])
index

plt.figure(figsize=(12,5))
plt.bar(index, df['total_servings'])
plt.xlim(0,195)

for i in index:
  plt.text(i, df['total_servings'][i], df['total_servings'][i], ha='center' )
plt.show()
```
