```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# x 축 : 지원번호 / y축 : 국어점수

x = df.index
y = df['국어']
e = df['영어']
m = df['수학']
# 그래프 크기 조정 / 그래프 그래픽 해상도 설정(dpi) : 해상도가 클 수록 이미지도 커짐
plt.figure(figsize=(10,5), dpi=300, facecolor='yellow')
plt.plot(x,y, label='국어점수', color='red', marker='s', mfc='yellow', mec='blue', ls='--')
plt.plot(x,e, label='영어점수')
plt.plot(x,m, label='수학점수')
plt.title('지원자 국어점수')
plt.xlabel('지원번호')
plt.ylabel('국어점수')
plt.legend(loc='lower right')
# 그래프 파일로 저장
# plt.savefig('graph2.png', dpi=300)
plt.show()
```
