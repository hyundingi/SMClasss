```python
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx')
```

grid : 안내선 출력

```python
plt.plot(df['이름'], df['키'])
plt.xticks(rotation=45)

# 그래프에 그리드 넣기 axis : x, y 축으로 선이 그려짐
plt.grid(axis='y', ls='--', lw=2, color='purple', alpha=0.2)
plt.grid(axis='x', ls='--', alpha=0.2)

# 그래프 값 추가
for i in range(df.shape[0]):
  plt.text(df['이름'][i], df['키'][i]+1, df['키'][i], ha='center')

plt.show()
```
