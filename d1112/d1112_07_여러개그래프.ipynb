```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]  # 리스트, 튜플, DataFrame 데이터 - 리스트 형태, Serise
y = [2,4,8]  # SM매출액
y2 = [5,1,3] # LG매출액
y3 = [1,2,5] # SK매출액
plt.figure(figsize=(10,5))

plt.plot(x,y, label='SM매출액', marker='o', linestyle='--')
plt.plot(x,y2, label='LG매출액', marker='s', linestyle='-.')
plt.plot(x,y3, label='SK매출액', marker='^', linestyle=':')

## 그래프 내 값 출력
for i in range(3):
  plt.text(x[i], y[i]+0.2, y[i], ha='center')
  plt.text(x[i], y2[i]+0.2, y[i], ha='center')
  plt.text(x[i], y3[i]+0.2, y[i], ha='center')

plt.title('기업별 매출액')
plt.legend(ncol=3) # 범례출력 기본 : 세로출력, ncol=3 : 가로출력 
plt.show()
```
