```python
import pandas as pd

df = pd.read_excel('score.xlsx', index_col='지원번호')


# 학교 이름 변경 : 구로고 > 가산고 / 디지털고 > 단지고
# SW특기 > 소문자로 변경
# 4번 SW특기 > Django 
# 5번 sw특기 > Python 
# 합계 컬럼 추가
# 결과컬럼 추가 270 이상 > pass
```

```python
df['학교'].replace({'구로고':'가산고','디지털고':'단지고'})
df['SW특기'].str.lower()
df.loc['4번', 'SW특기'] = 'Django'
df.loc['5번', 'SW특기'] = 'Python'

df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['결과'] = 'Fail'

# filt = df['합계'] >= 40000
# df.loc[filt, '결과'] = 'pass'
df.loc[df['합계'] >= 400, '결과'] = 'pass'
df
```

```python
# 문자열 =  문자열 + 문자열  가능
df['학교'] = df['학교']+'등학교'
df
```
