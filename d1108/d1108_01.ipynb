문자열 함수 : startswith, contains, replace, lower, upper

```python
# 해당 값으로 시작되는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains  : 해당 값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt, ['이름','학교','키']]

# isin() : 포함이 되어있는지 확인 / 괄호 안에는 리스트만 올 수 있음
langs = ['Python','Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower() : 해당값을 모두 소문자로 변환
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# 포함이 되어 있으면 모두 검색
filter = df['SW특기'].str.lower().isin(['j','v'])
df[filt]

# NaN 값도 처리해야 값이 나옴
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]

# 해당컬럼에 값을 넣으면 컬럼의 값이 수정됨
df['SW특기'] = df['SW특기'].str.lower()
```

null값 처리 : fillna [널 값 채우기] , dropna [널 값 삭제하기]

```python
import numpy as np

# fillna : inplace=True 해야 적용됨
# 모든 컬럼의 NaN 데이터를 모두 변경함
df.fillna('데이터 없음', inplace=True)
df
```

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
# df['학교'] = np.nan   # 학교 컬럼에 NaN 데이터 넣기
df.fillna(0)
```

```python
## dropna  [디폴트] index, any
## axis = 'index','columns' / 축방향 선택하여 삭제
#   how = 'any', 'all'  / any: 한개라도 있으면 모든 줄이나 열 삭제 , all : 줄 혹은 열의 모든 데이터가 nan 일 때 삭제
df.dropna()
df['학교'] = np.nan
df
df.dropna(axis='index', how='all')  # 모두 nan인 row가 없어서 삭제 불가
df.dropna(axis='columns', how='all') # 학교 컬럼이 모두 nan 이기에 삭제됨
df.dropna(axis='index', how='any') # 학교 컬럼이 모두 nan 이기에 모든 row삭제
df.dropna(axis='columns', how='any') # 학교, sw특기 컬럼에 nan이 있기 때문에 두 열 삭제

```
