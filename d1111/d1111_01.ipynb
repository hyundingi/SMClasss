```python
# 1. 요일 컬럼 추가
# 2. 0-6까지의 숫자를 랜덤으로 입력
# 3. 함수를 사용해서 0-일요일 ~
import random
def rand(value):
  value = random.randint(0,6)
  return str(value)

def yo1(value):
  if value == '0':
    return '일요일'
  elif value == '1':
    return '월요일'
  elif value == '2':
    return '화요일'
  elif value == '3':
    return '수요일'
  elif value == '4':
    return '목요일'
  elif value == '5':
    return '금요일'
  elif value == '6':
    return '토요일'
  
df['요일'] = df['요일'].apply(rand)
df['요일'] = df['요일'].apply(yo1)
df

```
