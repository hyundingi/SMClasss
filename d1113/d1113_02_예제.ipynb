```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
def cm_func(x):
  return str(x)+'cm'

# df['키'] : 배열타입에서 함수 호출할 때
df['키'].apply(cm_func)
```

```python
# 50점 아래 값 > 50으로 리턴해서 출력

def k_func(x):
  if x >= 50:
    return x
  return 50

# int, float, str 같은 단일 값은 함수명(데이터값) 으로 호출
for i in range(8):
  print(k_func(df['국어'][i]))
```

```python
def h_func(x):
  return int(x/2)

## 누적 막대 그래프

plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어'])
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'])

for i in range(8):
  plt.text(df['이름'][i], h_func(df['국어'][i]), df['국어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i]+h_func(df['영어'][i]), df['영어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i]+df['영어'][i]+h_func(df['수학'][i]), df['수학'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i]+df['영어'][i]+df['수학'][i]+7,df['국어'][i]+df['영어'][i]+df['수학'][i], ha='center')

plt.show()
```
