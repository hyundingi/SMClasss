```python
## 남자 인구현황 데이터
df = pd.read_excel('연령별인구현황_202410.xlsx', skiprows=3, usecols='B, E:Y', nrows=1, index_col='행정기관')

## 여자 인구현황 데이터
df2 = pd.read_excel('연령별인구현황_202410.xlsx', skiprows=3, usecols='B, AY:BS', nrows=1, index_col='행정기관')

x = df.columns
y = df.iloc[0].str.replace(',','').astype(int) // 1000
x2 = df2.columns
y2 = df2.iloc[0].str.replace(',','').astype(int) // 1000

plt.figure(figsize=(15,6))
plt.title('2024년 인구현황')
plt.barh(x,y, label='남자')
plt.barh(x,-y2, label='여자')
plt.xticks(rotation=45)
plt.legend(ncol=2)
plt.show()

# 
```
