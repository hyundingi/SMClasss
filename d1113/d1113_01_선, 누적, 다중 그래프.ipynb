선그래프

```python
plt.title('학생성적그래프')
plt.plot(df.index, df['국어'], label='국어')
plt.plot(df.index, df['영어'], label='영어')
plt.plot(df.index, df['수학'], label='수학')


for i in range(8):
  plt.text(df.index[i], df['국어'][i]+2, df['국어'][i], ha='center')
  plt.text(df.index[i], df['영어'][i]+2, df['영어'][i], ha='center')
  plt.text(df.index[i], df['수학'][i]+2, df['수학'][i], ha='center')

# 범례
plt.legend()
plt.show()
```

누적막대그래프

```python
plt.figure(figsize=(7,4))

plt.title('학생성적그래프')
plt.bar(df['이름'], df['국어'], width=0.6, label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], width=0.6, label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'], width=0.6, label='수학')

plt.legend()

for i in range(8):
  plt.text(df['이름'][i], df['국어'][i]/2, df['국어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i]+df['영어'][i]/2, df['영어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i]+df['영어'][i]+df['수학'][i]/2, df['수학'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i]+df['영어'][i]+df['수학'][i]+5, df['국어'][i]+df['영어'][i]+df['수학'][i], ha='center')

plt.show()
```

다중막대그래프

```python
idx = np.arange(8)
w = 0.25

plt.bar(idx-w, df['국어'], width=w)
plt.bar(idx, df['영어'], width=w)
plt.bar(idx+w, df['수학'], width=w)

plt.show()
```
