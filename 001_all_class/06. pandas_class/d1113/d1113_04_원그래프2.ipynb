```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
values = [30,25,20,13,10,2]    # 100 퍼센트
explode = [0,0.1,0,0,0,0]
wedgeprops = {'width':0.6}
labels = ['python','java','javascript','oracle','html', 'etc']
colors = ['y','g','peachpuff','bisque','plum','m']
wedgeprops = {'width':0.6, 'edgecolor':'white', 'linewidth':2}

## 데이터 값 : 2% 이하의 값은 출력되지 않도록 함수 적용
## pct : values의 퍼센트 값 생성
# def custom_autopct(pct):
#   if pct >= 10:
#     # return ('%.0f%%' % pct)
#     return ('{:.0f}%'.format(pct))
#   return ''

def custom_autopct(pct):
  return f'{pct:.0f}%' if pct >= 10 else ''


#  wedgeprops=도넛형 만들기 (중간에 구멍 만들기) / edgecolor 테두리 / linewidth 테두리 두께
plt.pie(values, labels=labels, colors=colors, autopct=custom_autopct, wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```
