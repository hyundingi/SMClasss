```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False

data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
## 영화 관객 수를 막대그래프로 출력
## 평균 = 2024년 평균 관객수 / 평균 평점
## 평균 막대그래프 추가, 평균 선을 그어서 .. 표시 

# 평균 row 추가 
df.loc[8] = ['평균',2024, df['관객 수'].mean(), df['평점'].mean()]

plt.figure(figsize=(10,5))
plt.bar(df['영화'], df['관객 수'])[len(df)-1].set_color('red')
plt.grid(axis='y', ls='--', alpha=0.2)
plt.xticks(rotation=-45)
plt.plot(df['영화'], [df.loc[8, '관객 수']]*9, ls=':', color='black')

plt.text(df['영화'][0], df['관객 수'][0], int(df['관객 수'][0]), ha='center')
plt.ylim(1000,1800)
# plt.show()
```
