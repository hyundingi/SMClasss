```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('chipotle.tsv', sep='\t')
df
```

```python
# 메뉴별 주문개수 5개만 뽑아서 막대그래프로 출력
# 순서 상관 없이, head(5)

menu_df = df.groupby('item_name')['order_id'].count().head(10)
menu_df['평균'] = df.groupby('item_name')['order_id'].count().mean()

plt.figure(figsize=(15,5))
plt.grid(ls='--', alpha=0.5)
plt.bar(menu_df.index, menu_df.values)[len(menu_df)-1].set_color('red')
plt.plot(menu_df.index, [menu_df['평균']]*11, ls=':', color='green')
for i in range(len(menu_df)):
  plt.text(menu_df.index[i], menu_df.values[i]+2, menu_df.values[i], ha='center')
plt.xticks(rotation=-45)
plt.show()
```
