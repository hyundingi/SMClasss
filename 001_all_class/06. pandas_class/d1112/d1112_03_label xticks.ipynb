```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# x축, y축 의 데이터
x = [-1,2,3]  # 리스트, 튜플, DataFrame 데이터 - 리스트 형태, Serise
y = [-2,-4,8]

## 선그래프 생성
plt.plot(x,y)

# 상단 타이틀 삽입
plt.title('라인그래프 Line', size='50')        

# 그래프 하단(x) 축 설명, color='' 글자색 변경, loc : 글자 위치(right, left, center)
plt.xlabel('X축 날짜', color='red', loc='left') 

# 그래프 좌측(y) 축 설명, loc : 글자 위치(top, bottom, center)
plt.ylabel('Y축 인원', color='green', loc='top')        

# 그래프 눈금 단위 표시
plt.xticks([1,2,3])
plt.yticks([2,4,8,10])


plt.show()
```
