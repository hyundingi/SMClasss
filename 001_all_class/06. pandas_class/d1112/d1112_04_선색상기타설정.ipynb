```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# x축, y축 의 데이터
x = [1,2,3]  # 리스트, 튜플, DataFrame 데이터 - 리스트 형태, Serise
y = [2,4,8]
yy = [1,8,3]

## 선그래프 생성
# linewidth : 선두께 , color : 선 색상, linestyle(ls) : 점선(:), 대쉬선(--), 점&선(-.) 
# marker : o,s,v,p 등 / markersize(ms), markeredgecolor, markerfacecolor(mfc), alpha
plt.plot(x,y, label='인원증가', color='pink', linewidth=5, marker='o', ms=10, markeredgecolor='red', mfc='yellow')
# plt.plot(x,yy, label='비용증가', marker='s', ms=10, alpha=0.4, ls='-.')

# 설정단축표시
# b: 선 색상, v: 마커, : 선 모양
plt.plot(x,yy, 'bv:', label='비용변동')

# 상단 타이틀 삽입
plt.title('라인그래프 Line')   

# 범례 : 그래프 선 설명문구/ 
# loc: 범례 위치 (upper,lower / right,left,center)
# plt.legend(loc='upper center')
# loc: 범례 위치 위치값으로 설정
plt.legend(loc=(0.3,0.1))
plt.show()
```
