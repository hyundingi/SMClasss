```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # win용
matplotlib.rcParams['font.size'] = 15                # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

bar 그래프 출력 /
라벨 변경 > 라벨 기울기 설정, width alpha 조절 가능

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
colors = ['r','g','b']
plt.figure(figsize=(5,3))  # 그래프 크기 조절

# 선그래프
plt.plot(label,values)

# 막대그래프
# color : 색상, width: 넓이 조절
# plt.bar(label, values, color='g')  한개의 색상 지정
plt.bar(label, values, color=colors, width=0.2, label='학생키') # 색상 여러개로 처리
plt.legend(loc='upper center')
# plt.legend(loc=(0.5,0.8))
plt.title('막대그래프')  # 상단 제목

plt.yticks([0,50,100,150,200,250])
plt.xticks(label, ['1번','2번','3번'], rotation=45)  # x라벨 변경

plt.show()
```

막대그래프 무늬 설정

```python
colors = ['red', 'green', 'blue']
bar = plt.bar(label, values, width=0.5, color = colors, alpha=0.5)

# 막대그래프 무늬 설정
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

범위 설정 xticks, yticks   -- 0부터 값이 표시됨
* 특정구간부터 범위설정 : xlim, ylim

```python
colors = ['red', 'green', 'blue']
bar = plt.bar(label, values, width=0.5, color = colors, alpha=0.5)
# plt.yticks([100,150,200]) # 범위 ㅣ정
plt.ylim(175,199) # 특정구간 범위 지정 
plt.show()
```

```python
colors = ['red', 'green', 'blue']
values = [180,187,184]
plt.barh(label,values, height=0.5, color=colors)
plt.xlim(175,190)
plt.show()
```
